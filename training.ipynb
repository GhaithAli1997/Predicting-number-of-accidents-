```python
#read the dataset then take the data until 2020
import pandas as pd
file_path = '/content/monatszahlen2307_verkehrsunfaelle_10_07_23_nosum.csv'
data = pd.read_csv(file_path)
filtered_data_untile_2020=data[data['JAHR'] <= 2020]
filtered_data_untile_2020
```

```python
#we have the column "MONAT" we must convert it to the month.
filtered_data_untile_2020['MONAT'] = filtered_data_untile_2020['MONAT'].astype(str)
filtered_data_untile_2020['MONAT'] = filtered_data_untile_2020['MONAT'].str[-2:]
filtered_data_untile_2020['MONAT'] = filtered_data_untile_2020['MONAT'].apply(lambda x: int(str(x)[-2:]))
filtered_data_untile_2020
```

```python
import matplotlib.pyplot as plt

# Group data by 'Category' and 'Year' to sum the 'Value' column
category_data = filtered_data_untile_2020.groupby(['JAHR','MONATSZAHL' ])['WERT'].sum().unstack()

# Plotting the historical number of accidents per category
category_data.plot(kind='line', figsize=(12, 6))
plt.title('Historical Number of Accidents per Category')
plt.xlabel('JAHR')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
category_data = filtered_data_untile_2020.groupby(['MONATSZAHL','JAHR' ])['WERT'].sum().unstack()
category_data.plot(kind='bar', figsize=(12, 6))
plt.title('Historical Number of Accidents per Category')
plt.xlabel('Category')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
#take the rows when the "MONATSZAHL" is "Alkoholunfälle" and "AUSPRAEGUNG" is "insgesamt"
filtered_data_untile_2020 = filtered_data_untile_2020[(filtered_data_untile_2020['MONATSZAHL'] == 'Alkoholunfälle') & (filtered_data_untile_2020['AUSPRAEGUNG'] == 'insgesamt')]
filtered_data_untile_2020
```

```python
#take the important colomuns.
filtered_data_untile_2020=filtered_data_untile_2020.iloc[:, 2:5]
filtered_data_untile_2020
```

```python
#define the model "Linear Regression" and  define the "X" and y which they are the colomuns 'JAHR' and 'MONAT' for "X" and WERT for y.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
model = LinearRegression()
X = filtered_data_untile_2020.drop('WERT', axis=1)
y = filtered_data_untile_2020['WERT']
model.fit(X, y)

```

```python
#test the value 2021-1
jan_2021 =pd.DataFrame({ 'JAHR': [2021], 'MONAT': [1]})
prediction = model.predict(jan_2021)

print(f"Predicted number of accidents for January 2021: {prediction[0]}")
```

```python
#save the model
import pickle
with open('Digital_prodact.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Load the saved model
with open('/content/Digital_prodact.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
jan_2021 =pd.DataFrame({ 'JAHR': [2021], 'MONAT': [1]})
prediction = loaded_model.predict(jan_2021)

print(f"Predicted number of accidents for January 2021: {prediction[0]}")


```
